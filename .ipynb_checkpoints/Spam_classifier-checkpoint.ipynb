```python
stopwords = nltk.corpus.stopwords.words('english')
word_list = nltk.corpus.words.words()

def extract_misspelled(x):
    words = nltk.word_tokenize(x)
    # Remove punctuation and convert to lowercase
    words =[word.lower() for word in words if word.isalpha()]
    # Remove stopwords
    words = [word for word in words if word not in stopwords]
    misspelled_words = [word for word in words if word not in word_list]
    #print(misspelled_words)
    return len(misspelled_words)
```

Data Preprocessing

```python
import string
string.punctuation
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
ps.stem('dancing')
```

```python
def transform_text(text):
    text = text.lower() #lower the text
    text = word_tokenize(text) # convert to words
    y = []
    for i in text: #remove special characters and punctuations
        if i.isalnum():
            y.append(i)
   
    text = y.copy() #remove stopwords and special characters
    y.clear()
    #rint(text)
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    
    text = y.copy() # remove stem words
    y.clear()
    #rint(text)
    for i in text:
        y.append(ps.stem(i))
    
    return " ".join(y) # return as a string
```

```python
(df['text'][3],transform_text(df['text'][3]))
```

```python
df['transformed_text'] = df['text'].apply(transform_text)
```

Our data is imbalanced
i.e there are less spam messages than non spam messages, so we shouldn't consider accuracy is the best metric to validate model's performance.

The problem is determining if a message is spam or not spam.

The assumptions are that 0 means ham(negative) and 1 means spam(Positive)

If the model predicts a spam message as ham (False Negative), There won't be much cost, but 

However,If the model predicts a ham message as spam (False positive), Use will loose important mail, it is important to avoid falsely classifying legitimate emails as spam.
 
To achieve less damage (low false positive rate), we must aim for a high precision value.

Precision is calculated as True Positives / (True Positives + False Positives).

```python
#MultinomialNB
mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
```

```python
df = pd.read_csv('df_processed.csv')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
X = df.transformed_text
```

```python
y = df['target']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(X).toarray()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
lr = LogisticRegression()
svc = SVC(kernel ='sigmoid')
mnb = MultinomialNB()
dtc = DecisionTreeClassifier()
KNN = KNeighborsClassifier()
rfc = RandomForestClassifier(random_state = 36)
abc = AdaBoostClassifier(random_state = 36)
bc = BaggingClassifier(random_state = 36)
etc = ExtraTreesClassifier(random_state = 36)
gbc = GradientBoostingClassifier(random_state = 36)
xgb = XGBClassifier(random_state = 36)
```

```python
clfs = {
    'lr' : lr,
    'scv' : svc,
    'mnb' : mnb,
    'dtc' : dtc,
    'knn' : KNN,
    'rfc' : rfc,
    'abc' : abc,
    'bc' : bc,
    'etc' : etc,
    'gbc' : gbc,
    'xgb' : xgb
}
```

```python
def model_performance(clf,X_train,X_test,y_train,y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    return np.round(accuracy_score(y_test,y_pred),3),np.round(precision_score(y_test,y_pred),3)
    
```

```python
model_performance(lr,X_train,X_test,y_train,y_test)
```

```python
accuracy_scores = []
precision_scores =[]

for name,clf in clfs.items():
    current_accuracy,current_precision = model_performance(clf,X_train,X_test,y_train,y_test)
    
    print('For ',name)
    print('Accuracy ',current_accuracy)
    print('Precision ',current_precision)
     
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    
```

```python
df_clfs = pd.DataFrame({'Classifier':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores}).sort_values(by ='Precision',ascending = False)
```

```python
df_clfs
```

```python
df_clfs = pd.melt(df_clfs,id_vars='Classifier')
```

Part 5 : improving model performance

1. change the max_features parameter of tfidf 
2. also calculate theif fit and predict times
3. change n_estimators=50 parameter in tree models

```python
from time import time
```

```python
df.isnull().sum()
```

```python
X = df['transformed_text']
```

```python
y
```

```python
tfidf = TfidfVectorizer(max_features=3000)
X = tfidf.fit_transform(X).toarray()
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
lr = LogisticRegression()
svc = SVC(kernel ='sigmoid',probability=True)
mnb = MultinomialNB()
dtc = DecisionTreeClassifier()
KNN = KNeighborsClassifier()
rfc = RandomForestClassifier(n_estimators=50, random_state = 36)
abc = AdaBoostClassifier(n_estimators=50,random_state = 36)
bc = BaggingClassifier(n_estimators=50,random_state = 36)
etc = ExtraTreesClassifier(n_estimators=50,random_state = 36)
gbc = GradientBoostingClassifier(n_estimators=50,random_state = 36)
xgb = XGBClassifier(n_estimators=50,random_state = 36)
```

```python
def model_performance(clf,X_train,X_test,y_train,y_test):
    t0=time()
    clf.fit(X_train,y_train)
    fit_time = round(time()-t0, 3)
    t1=time()
    y_pred = clf.predict(X_test)
    predict_time = round(time()-t1, 3)
    return np.round(accuracy_score(y_test,y_pred),3),np.round(precision_score(y_test,y_pred),3),fit_time,predict_time
```

```python
model_performance(mnb,X_train,X_test,y_train,y_test)
```

```python
accuracy_scores = []
precision_scores =[]
fit_times =[]
predict_times = []

for name,clf in clfs.items():
    current_accuracy,current_precision,current_fit_time,current_pred_time = model_performance(clf,X_train,X_test,y_train,y_test)
    
    print('For ',name)
    print('Accuracy ',current_accuracy)
    print('Precision ',current_precision)
     
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    fit_times.append(current_fit_time)
    predict_times.append(current_pred_time)
```

```python
df_clfs.columns
```

```python
#pivottable pf df_clfs
df_clfs = pd.pivot_table(df_clfs,values='value',columns='variable',index='Classifier').reset_index().sort_values(by = 'Precision',ascending=False)
```

```python
df_clfs 
```

```python
df_clfs_2 = pd.DataFrame({'Classifier':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores,'fit_times':fit_times,'predict_times':predict_times}).sort_values(by ='Precision',ascending = False)
```

```python
df_clfs_2.sort_values(by='Precision',ascending=False)
```

```python
df_clfs_final = df_clfs.merge(df_clfs_2,on ='Classifier')
df_clfs_final 
```
